```python
import pandas as pd
import numpy as np

input_dir = "out03_age_prediction_hc2_stdz_age_reverse_fa_scale_thresh0.02"

model = 'km4'

# read age prediction and behavioral data:
scatter_data = pd.read_csv(input_dir + '/out04_scatter_data_ridge_cluster_gmm_' + model + '_fit_brain_age.csv', index_col = 0)
scatter_data_hc = scatter_data.loc[scatter_data['group']=='HC',]

scatter_data_wide = scatter_data_hc.pivot(index='SUBJID',columns='cluster', 
                                       values = ['brain age', 'age gap', 'brain age fit intersex', 'age gap corr'])

scatter_data_wide.columns = scatter_data_wide.columns.map('|'.join).str.strip('|')

behav_data = pd.read_csv('out01_robustreg_behav_hc2_stdz_age_reverse_fa/out01_behav_cog_hc_raw.csv', index_col = 0)
behav_data.drop(columns = 'names', inplace = True)

data_merge = scatter_data_wide.merge(behav_data, on = 'SUBJID')
data_merge['Sex'].replace({'F':0,'M':1},inplace=True)
```

```python
Y_brain_age = data_merge[['brain age|1.0', 'brain age|2.0', 'brain age|3.0', 'brain age|4.0']].values
Y_age_gap = data_merge[['age gap corr|1.0', 'age gap corr|2.0', 'age gap corr|3.0', 'age gap corr|4.0']].values

y_brain_age_all = data_merge[['brain age|all']].values
y_chro_age = data_merge[['age_at_cnb']].values

X_behav = data_merge[behav_data.columns[3:]].values
X_sex = np.vstack(data_merge['Sex'].values)
```

```python
X_behav
```

```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

X_scaled = preprocessing.scale(X_behav)

Y = Y_brain_age
X = np.hstack((X_sex, X_scaled))

rows_idx = ~np.any(np.isnan(X), axis = 1)
R2_brain_age = []

X_train, X_test, y_train, y_test = train_test_split(X[rows_idx,:], Y[rows_idx,:], test_size=.5, random_state=42)

for n_comps in range(1,10):

    pls = PLSRegression(n_components=n_comps)
    pls.fit(X_train, y_train)

    R2_brain_age.append(pls.score(X_test, y_test))
    
print(R2_brain_age)
```

```python
Y = y_brain_age_all
X = np.hstack((X_sex, X_scaled))

rows_idx = ~np.any(np.isnan(X), axis = 1)

R2_brain_age_all = []
X_train, X_test, y_train, y_test = train_test_split(X[rows_idx,:], Y[rows_idx,:], test_size=0.5, random_state=42)

for n_comps in range(1,10):

    pls = PLSRegression(n_components=n_comps)
    pls.fit(X_train, y_train)

    R2_brain_age_all.append(pls.score(X_test, y_test))
    
print(R2_brain_age_all)
```
